Programa para calcular o salário

```python
print("Bem vindo ao programa de cálculo do valor por hora trabalhada!")
valor_por_hora= float(input("Digite sua remuneração por horas: "))
horas_trabalhadas= float(input("Digite o total de horas trabalhadas no mês: "))
salario= valor_por_hora * horas_trabalhadas
imposto_de_renda= salario * 0.11
inss= (salario - imposto_de_renda) * 0.08
sindicato= (salario - imposto_de_renda - inss) * 0.05
desconto= imposto_de_renda + inss + sindicato
salario_liquido= salario - desconto
print("======================")
print(f"1: O salário bruto é de: R$ {salario:.2f} ")
print(f"2: Total do IR é de: R$ {imposto_de_renda:.2f}")
print(f"3: Total do INSS é de: R$ {inss:.2f}")
print(f"4: Total do sindicado é de: R$ {sindicato:.2f}")
print(f"5: Total dos descontos: R$ {desconto:.2f}")
print(f"6: Salário líquido: R$ {salario_liquido:.2f} ")
```
